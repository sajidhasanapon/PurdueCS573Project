```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('vgg_training.log')
```

```python
data.head()
```

```python
plt.plot(data['epoch'],data['acc']*100,label='Training')
plt.plot(data['epoch'],data['val_acc']*100,'-*', color='orange',label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('VGG16_transfer_learning.pdf',format='pdf')
```
